```python
%config InlineBackend.rc = {"figure.dpi": 72, "figure.figsize": (6.0, 4.0)}
%matplotlib ipympl

import abtem
import ase
import matplotlib.pyplot as plt
import numpy as np
```

```python
atoms_unit = ase.build.bulk("Au", cubic=True)
atoms = atoms_unit * (1, 1, 60)

# fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))
# abtem.show_atoms(atoms, ax=ax1, title="Beam view")
# abtem.show_atoms(atoms, ax=ax2, plane="xz", title="Side view", linewidth=0);
```

```python
from abtem.tilt import precession_tilts

wave = abtem.PlaneWave(energy=200e3)

wave.tilt = precession_tilts(50, 30)

potential = abtem.Potential(atoms, sampling=0.03, exit_planes=1)

exit_wave = wave.multislice(potential)

exit_wave.compute()
```

```python
spots = exit_wave[1:].diffraction_patterns().index_diffraction_spots(
    atoms_unit
).block_direct()
```

```python
spots.shape
```

```python
metadata = abtem.core.axes.NonLinearAxis(label="azimuth", values=np.linspace(0, 2*np.pi, 30, endpoint=False), units="rad")
spots.set_ensemble_axes_metadata(metadata, axis=1)
```

```python
spots.show(interact=True, scale=0.5)
```
